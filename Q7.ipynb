```python
import matplotlib.pyplot as plt
import random
import numpy as np

class MonteCarloSim:
    
    def __init__(self,N):
        
        self.ParentNucleus = []
        self.DaughterNucleus = []
        self.HalfLife = 4.5*60
        self.DecayConst = 1/self.HalfLife    
        self.probToDecay = 0.01
        
        self.N = N
        self.Np = N 
        self.Nd = 0
        
        self.time = np.arange(0, 100, 1)
    
    def simulate(self):
        
        for t in self.time:
            self.ParentNucleus.append(self.Np)
            self.DaughterNucleus.append(self.Nd)
            
            decay = 0
            
            for n in range(self.Np):
                p = random.random()
                if p < self.probToDecay:
                    decay+=1
            
            self.Np -=decay
            self.Nd +=decay
            
    def plotfig(self):

        plt.plot(self.ParentNucleus, linestyle='-', color='red', label='Parent Nucleus')
        plt.plot(self.DaughterNucleus, linestyle='-', color='blue', label='Daughter Nucleus')
        plt.legend()
        plt.xlabel('t', fontsize=14)
        plt.ylabel('N(t)', fontsize=14)
        plt.show()
```

```python
m = MonteCarloSim(100)
m.simulate()
m.plotfig()
```

```python
m = MonteCarloSim(500)
m.simulate()
m.plotfig()
```
